```python
#Function for variable rate calculations
def amount_vr(list_rates, seed_capital):
  amount = seed_capital
  for rates in list_rates:
    amount *= (1 + rates)
  return amount
```

```python
#Applying variable rate function
list_rates = [0.008, 0.003, 0.002]
seed_capital = 100000
amount = amount_vr(list_rates, seed_capital)
print(f'The amount is ${round(amount, 2)}')
```

```python
#Function for finding the accrued fee
def accrued_fee(list_rates):
  import numpy
  list_values=[1 + rate for rate in list_rates]
  fee = (numpy.prod(list_values) - 1)
  return fee
```
