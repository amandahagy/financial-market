```python
import handcalcs.render as render
from handcalcs import handcalc
```

```python
@handcalc(jupyter_display=True)
def proportional_fees(interest_rate1, period1, period2):
  interest_rate2 = period1 * interest_rate1 / period2
  return interest_rate2
```

```python
#Applying proportional fees function
interest_rate1 = 0.06
period1 = 1 #year
period2 = 4 #trimester
interest_rate2 = proportional_fees(interest_rate1, period1, period2)
print(interest_rate2)
```

```python
@handcalc(jupyter_display=True)
def equivalent_fees(interest_rate1, period1, period2):
  interest_rate2 = (1 + interest_rate1) ** (period1 / period2) - 1
  return interest_rate2
```

```python
#Applying equivalent fees function
interest_rate1 = 0.065
period1 = 1 #year
period2 = 4 #trimester
interest_rate2 = equivalent_fees(interest_rate1, period1, period2)
print({round(interest_rate2, 2)})
```

```python
#Applying Brazilian Central Bank's over rate
interest_rate1 = 0.03 #per working year
period1 = 1 #working year
period2 = 252 #working days
interest_rate2 = equivalent_fees(interest_rate1, period1, period2)
print(f'The equivalent fees are {round(interest_rate2, 6) * 100}% per working day.')
```
