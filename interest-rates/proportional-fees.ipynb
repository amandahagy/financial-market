```python
import handcalcs.render as render
from handcalcs import handcalc
```

```python
@handcalc(jupyter_display=True)
def proportional_fees(interest_rate1, period1, period2):
  interest_rate2 = period1 * interest_rate1 / period2
  return interest_rate2
```

```python
#Applying proportional fees function
interest_rate1 = 0.06
period1 = 1 #year
period2 = 4 #trimester
interest_rate2 = proportional_fees(interest_rate1, period1, period2)
print(interest_rate2)
```
