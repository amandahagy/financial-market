```python
# Function for simple discrete capitalization
def amount_sd(main_capital, interest_rate, period):
  SD = (1 + period * interest_rate) * main_capital
  return SD
```

```python
# Applying function for simple discrete capitalization
main_capital = 20000
period = 5
interest_rate = 0.07
SD = amount_sd(main_capital, interest_rate, period)
```

```python
# Function for compound discrete capitalization
def amount_cd(main_capital, interest_rate, period):
  CD = main_capital * (1 + interest_rate) ** period
  return CD
```

```python
# Applying function for compound discrete capitalization
main_capital = 20000
period = 4
interest_rate = 0.0065
CD = amount_cd(main_capital, interest_rate, period)
```

```python
# Comparison between simple and compound discrete capitalization
main_capital = 10000
interest_rate = 0.071
years = [0.5, 1, 1.5, 10]
SD = [amount_sd(main_capital, interest_rate, period) for period in years]
CD = [amount_cd(main_capital, interest_rate, period) for period in years]
difference = [x - y for x,y in zip(SD,CD)]
print(difference)
```

```python
#Graphic: simple discrete capitalization vs. compound discrete capitalization
import matplotlib.pyplot
matplotlib.pyplot.plot(years, SD, color = 'b', label = 'Simple Discrete Capitalization')
matplotlib.pyplot.plot(years, CD, color = 'r', label = 'Compound Discrete Capitalization')
matplotlib.pyplot.grid(True)
matplotlib.pyplot.xlabel('Years')
matplotlib.pyplot.ylabel('$')
matplotlib.pyplot.title('Simple vs. Compound')
matplotlib.pyplot.legend();
```
