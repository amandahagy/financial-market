```python
#Function for calculating average interest rates
def average_rate(list_rates):
  import numpy
  list_values = [1 + value for value in list_rates]
  rate = ((numpy.prod(list_values)) ** (1 / len(list_values)) - 1)
  return rate
```

```python
#Applying average interest rate function
list_rates = [0.005, 0.032, -0.025, 0.057] #4 monthly interest rates
rate = average_rate(list_rates)
print(f'The average return was {round(rate, 4)}% per month.')
```

```python
#Fisher Equation function
def fisher_function (effective_rate, inflation_rate, real_rate):
  if effective_rate == 0:
    return ((1 + inflation_rate) * (1 + real_rate) - 1)
  elif inflation_rate == 0:
    return ((1 + effective_rate) / (1 + real_rate) - 1)
  elif real_rate == 0:
    return ((1 + effective_rate) / (1 + inflation_rate) - 1)
```

```python
#Example for adding inflation to the interest rate
effective_rate = 0.015 #per month
inflation_rate = 0.005 #per month
real_rate = fisher_function (effective_rate, inflation_rate, 0) #set the value of what you want to find (in this case, the real interest rate) as 0
print(f'If we consider the inflation, the real return is {round((real_rate * 100), 3)}% per month.')
```

```python
#Applying Fisher Equation function for finding the effective interest rate
inflation_rate = 0.005 #per month
real_rate = 0.00995 #per month
effective_rate = fisher_function(0, inflation_rate, real_rate)
print(f'The effective rate is {round((effective_rate * 100), 3)}% per month.')
```

```python
#Finding the inflation rate by applying Fisher Equation function
effective_rate = 0.015 #per month
real_rate = 0.00995 #per month
inflation_rate = fisher_function(effective_rate, 0, real_rate)
print(f'Inflation rate: {round((inflation_rate * 100), 3)}% per month.')
```
