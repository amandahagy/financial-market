```python
#Function for calculating average interest rates
def average_rate(list_rates):
  import numpy
  list_values = [1 + value for value in list_rates]
  rate = ((numpy.prod(list_values)) ** (1 / len(list_values)) - 1)
  return rate
```

```python
#Applying average interest rate function
list_rates = [0.005, 0.032, -0.025, 0.057] #4 monthly interest rates
rate = average_rate(list_rates)
print(f'The average return was {round(rate, 4)}% per month.')
```

```python
#Fisher Equation function
def add_inflation (effective_rate, inflation_rate, real_rate):
  if effective_rate == 0:
    return ((1 + inflation_rate) * (1 + real_rate) - 1)
  elif inflation_rate == 0:
    return ((1 + effective_rate) / (1 + real_rate) - 1)
  elif real_rate == 0:
    return ((1 + effective_rate) / (1 + inflation_rate) - 1)
```
